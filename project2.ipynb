```python
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
file_names = ["003_2006_2080_352_360.nc", "004_2006_2080_352_360.nc", "005_2006_2080_352_360.nc", 
              "006_2006_2080_352_360.nc", "007_2006_2080_352_360.nc", "008_2006_2080_352_360.nc"]
```

```python
for file_name in file_names:
    ds = xr.open_dataset(file_name)
    ds["TREFMXAV_U"].mean(dim=["time"]).plot()
    plt.show()  # Show the plot
    ds = ds.to_dataframe()
    ds.to_csv(file_name + '.csv')
    print(ds.describe())
```

```python
# Combine data files
data = xr.open_mfdataset(file_names, concat_dim=['time'], combine='nested')
data
```

```python
# Filtering data for the Manchester area
lon = 357.5
lat = 53.246075

data_manchester = data.sel(lon=lon, lat=lat, method='nearest')
```

```python
# Extracting xarray time coordinates
time_data = data_manchester['time'].values

# Converting time coordinates to datetime format
datetime_data = pd.to_datetime([str(t) for t in time_data])
```

```python
# Replacing time coordinates of xarray data
data_manchester = data_manchester.assign_coords(time=datetime_data)
# Converting xarray data to DataFrame
df = data_manchester.to_dataframe().reset_index()

# Ensuring the time column is in datetime format
df['time'] = pd.to_datetime(df['time'])
```

```python
# Extracting time-related features
df['year'] = df['time'].dt.year
df['month'] = df['time'].dt.month
df['quarter'] = df['time'].dt.quarter
```

```python
df.head()
```

```python
# Checking for missing values in the dataset
null_counts = df.isnull().sum()

# Outputting the results of missing value statistics
print(null_counts)
```

```python
df.describe()
```

```python
negative_values_columns = ['FLNS', 'PRECT', 'PRSN', 'UBOT', 'VBOT']

# Iterate over each column containing negative values and output the corresponding data
for column_name in negative_values_columns:
    # Find row indices containing negative values
    negative_row_indices = df.index[df[column_name] < 0]
    # Extract data from these rows and save as a CSV file
    negative_rows_data = df.loc[negative_row_indices]
    filename = f'negative_values_{column_name}.csv'
    negative_rows_data.to_csv(filename, index=False)
```

```python
# Delete rows containing negative values in the 'FLNS', 'PRECT' column
df = df[df['FLNS'] >= 0]
df = df[df['PRECT'] >= 0]
df.describe()
```

```python
# View the distribution of negative values in the 'PRSN' column
# Load negative value data
data_PRSN = pd.read_csv(f'negative_values_PRSN.csv')
data_PRSN['time'] = pd.to_datetime(data_PRSN['time'])
plt.figure(figsize=(30, 6))
plt.plot(data_PRSN['time'], data_PRSN['PRSN'], marker='o', linestyle='-')
plt.title(f'PRSN vs Time')
plt.xlabel('Time')
plt.ylabel('PRSN')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.interpolate import interp1d

# Save the original data
original_data = df.copy()

# Replace negative values with NaN
df.loc[df['PRSN'] < 0, 'PRSN'] = np.nan

# Perform linear interpolation
df['PRSN'] = df['PRSN'].interpolate(method='linear')

# Save the processed data
df.to_csv('interpolated_data.csv', index=False)
```

```python
# Checking if there are still NaN values in the data after interpolation
nan_count_after_interpolation = df['PRSN'].isnull().sum()
print(f"NaN values after interpolation: {nan_count_after_interpolation}")
```

```python
# View the effect of interpolation
# Plot scatter plot for comparison of fitting degree
plt.figure(figsize=(10, 6))
plt.scatter(original_data['PRSN'], df['PRSN'])
plt.xlabel('Original PRSN')
plt.ylabel('Interpolated PRSN')
plt.title('Comparison of Original and Interpolated Data')
plt.grid(True)
plt.show()

# Calculate standard deviation or variance to measure the smoothness of the data
original_std = original_data['PRSN'].std()
interpolated_std = df['PRSN'].std()
print(f"Original Data Standard Deviation: {original_std}")
print(f"Interpolated Data Standard Deviation: {interpolated_std}")

# Calculate correlation coefficient to measure the fitting degree of the data
# Calculate Pearson correlation coefficient
pearson_corr = original_data['PRSN'].corr(df['PRSN'], method='pearson')

# Calculate Spearman correlation coefficient
spearman_corr = original_data['PRSN'].corr(df['PRSN'], method='spearman')

print(f"Pearson's Correlation Coefficient: {pearson_corr}")
print(f"Spearman's Correlation Coefficient: {spearman_corr}")
```

```python
# Climate Feature Analysis
# Define variables to analyze
variables = ['TREFMXAV_U', 'FLNS', 'FSNS', 'PRECT', 'PRSN', 'QBOT', 'TREFHT', 'UBOT', 'VBOT']

# Calculate monthly and yearly means for each variable
monthly_mean = df.groupby('month')[variables].mean()
yearly_mean = df.groupby('year')[variables].mean()
quarterly_mean = df.groupby('quarter')[variables].mean()

# Plot the distribution of mean values for each variable across months
for var in variables:
    plt.figure(figsize=(10, 6))
    plt.plot(monthly_mean.index, monthly_mean[var])
    plt.title(f'Monthly Mean {var}')
    plt.xlabel('Month')
    plt.ylabel('Mean Value')
    plt.tight_layout()
    plt.show()

# Plot the distribution of mean values for each variable across years
for var in variables:
    plt.figure(figsize=(15, 6))
    plt.plot(yearly_mean.index, yearly_mean[var])
    plt.title(f'Yearly Mean {var}')
    plt.xlabel('Year')
    plt.ylabel('Mean Value')
    plt.tight_layout()
    plt.show()

# Plot the distribution of mean values for each variable across quarters
for var in variables:
    plt.figure(figsize=(15, 6))
    plt.plot(quarterly_mean.index, quarterly_mean[var])
    plt.title(f'Quarterly Mean {var}')
    plt.xlabel('quarter')
    plt.ylabel('Mean Value')
    plt.tight_layout()
    plt.show()
```

```python
# Adding periodicity
# Converting month to sine and cosine
df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)  # 12 is the period
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
# Converting quarter to sine and cosine
df['quarter_sin'] = np.sin(2 * np.pi * df['quarter'] / 4)  # 4 is the period
df['quarter_cos'] = np.cos(2 * np.pi * df['quarter'] / 4)
```

```python
# Correlation Analysis
# Calculate Pearson correlation coefficient matrix
pearson_correlation_matrix = df.drop(columns=['time', 'lat', 'lon']).corr(method='pearson')

# Plot Pearson correlation coefficient heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pearson_correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Pearson Correlation Heatmap')
plt.tight_layout()
plt.savefig(f'Pearson Correlation Heatmap.pdf')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# Selecting training data
X_train = df[df['year'] < 2050].drop(columns=['time', 'lat', 'lon', 'TREFMXAV_U', 'TREFHT'])
y_train = df[df['year'] < 2050]['TREFMXAV_U']

# Selecting the test set using data from 2050 and beyond
X_test = df[df['year'] >= 2050].drop(columns=['time', 'lat', 'lon', 'TREFMXAV_U', 'TREFHT'])
y_test = df[df['year'] >= 2050]['TREFMXAV_U']
```

```python
# 1. Linear Regression Model
# Creating a pipeline for the model
pipeline_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LinearRegression())
])

# Defining parameter ranges
param_grid_lr = {
    'model__fit_intercept': [True, False]
}

# Using GridSearchCV for cross-validation and parameter tuning
grid_search_lr = GridSearchCV(pipeline_lr, param_grid_lr, cv=5, scoring='neg_mean_squared_error')
grid_search_lr.fit(X_train, y_train)

# Obtaining the best model
best_model_lr = grid_search_lr.best_estimator_

# Obtaining the best parameters
best_params_lr = grid_search_lr.best_params_

# Prediction and evaluation
y_pred_lr = best_model_lr.predict(X_test)
mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(mse_lr)
r2_lr = r2_score(y_test, y_pred_lr)
print(f'Best Parameters:  {best_params_lr}')
print(f'Best Mean Squared Error: {mse_lr}')
print(f'Best Root Mean Squared Error: {rmse_lr}')
print(f'Best R-squared: {r2_lr}')
```

```python
# 2. Random Forest Model
# Creating a pipeline for the model
pipeline_ram = Pipeline([
    ('scaler', StandardScaler()),
    ('model', RandomForestRegressor())
])

# Parameter ranges, selecting parameters for random forest
param_grid_ram = {
    'model__n_estimators': [100, 200, 300],
    'model__max_features': [None, 'sqrt', 'log2']
}

# Using GridSearchCV for cross-validation and selecting the best parameters
grid_search_ram = GridSearchCV(pipeline_ram, param_grid_ram, cv=5, scoring='neg_mean_squared_error')
grid_search_ram.fit(X_train, y_train)

# Obtaining the best model
best_model_ram = grid_search_ram.best_estimator_

# Obtaining the best parameters
best_params_ram = grid_search_ram.best_params_

# Prediction and evaluation
y_pred_ram = best_model_ram.predict(X_test)
mse_ram = mean_squared_error(y_test, y_pred_ram)
rmse_ram = np.sqrt(mean_squared_error(y_test, y_pred_ram))
r2_ram = r2_score(y_test, y_pred_ram)
print(f'Best Parameters:  {best_params_ram}')
print(f'Mean Squared Error: {mse_ram}')
print(f'Root Mean Squared Error: {rmse_ram}')
print(f'R-squared: {r2_ram}')
```

```python
# 3. Gradient Boosting Tree Model
# Creating a pipeline for the model
pipeline_gbt = Pipeline([
    ('scaler', StandardScaler()),
    ('model', GradientBoostingRegressor())
])

# Parameter ranges, selecting parameters for gradient boosting tree
param_grid_gbt = {
    'model__n_estimators': [100, 200, 300],
    'model__learning_rate': [0.05, 0.1, 0.2]
}

# Using GridSearchCV for cross-validation and selecting the best parameters
grid_search_gbt = GridSearchCV(pipeline_gbt, param_grid_gbt, cv=5, scoring='neg_mean_squared_error')
grid_search_gbt.fit(X_train, y_train)

# Obtaining the best model
best_model_gbt = grid_search_gbt.best_estimator_

# Obtaining the best parameters
best_params_gbt = grid_search_gbt.best_params_

# Prediction and evaluation
y_pred_gbt = best_model_gbt.predict(X_test)
mse_gbt = mean_squared_error(y_test, y_pred_gbt)
rmse_gbt = np.sqrt(mean_squared_error(y_test, y_pred_gbt))
r2_gbt = r2_score(y_test, y_pred_gbt)
print(f'Best Parameters:  {best_params_gbt}')
print(f'Mean Squared Error: {mse_gbt}')
print(f'Root Mean Squared Error: {rmse_gbt}')
print(f'R-squared: {r2_gbt}')
```

Reflective writing
1. Firstly, when merging datasets, it is crucial to include additional columns indicating the source of the data, enabling traceability of data contributions to the model. In future research, establishing a clear traceability process for data sources from different origins should be prioritized from the outset.
2. When analyzing data correlations, both highly and lowly correlated feature variables should be handled with caution. In this study, the temperature variables exhibited extremely high correlation, which, while advantageous for improving the accuracy of predicting maximum temperatures, significantly reduced the contributions of other variables. Therefore, it was challenging to discern the impact of other variables on maximum temperature. Further research could explore causality between other variables and maximum temperature after excluding temperature variables.
3. Persistently pursuing more complex models and parameters may not always yield favorable results. Initially opting for an SVM model and preparing numerous alternative parameters for grid search resulted in excessively long model training times and no conclusive results. Moreover, it could lead to overfitting. Hence, a balanced approach is necessary, weighing model complexity against precision and generalization ability.
4. This study did not address the effects of topographical features and human activities. Future research could investigate the interaction between global warming, topographical features, and human activities, and their combined impact on Manchester's temperature.
